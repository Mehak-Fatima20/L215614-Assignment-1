```python
file = pd.read_csv("/content/gdrive/MyDrive/StudentMarkSheet/CatData.csv")
```

```python
file
```

```python
file.dropna( axis=0, inplace = True)
```

```python
file
```

```python
#independent variable == Tail length
#dependent variable == Mass
```

```python
import numpy as np
from sklearn import linear_model

```

```python
file.columns
```

```python
file.info()
```

```python
file = file[["Tail Length (cm)", "Mass (kg)"]]
```

```python
print (file.head())
```

```python
plt.scatter(file["Tail Length (cm)"],file[ "Mass (kg)"])
plt.title("Length VS Mass")
plt.xlabel("TaiL")
plt.ylabel("Mass")
plt.show()
```

```python
#Taking 80% data for training
num = int (len(file)*0.8)
train = file [:num]
test = file [num:]
print ("Data: ",len(file))
print ("Train: ", len(train))
print ("Test: ", len(test))
```

```python
regr = linear_model.LinearRegression()
train_X = np.array(train[["Tail Length (cm)"]])
train_Y = np.array(train[["Mass (kg)"]])
regr.fit(train_X,train_Y)
Coefficients= regr.coef_
intercept = regr.intercept_
print ("Slope: ", Coefficients)
print("Intercept: ", intercept)
```

```python
plt.scatter(train["Tail Length (cm)"], train ["Mass (kg)"])
plt.plot(train_X,Coefficients[0]*train_X+intercept,color = "red")
plt.xlabel("Tail Length (cm)")
plt.ylabel("Mass (kg)")
plt.show 
```
